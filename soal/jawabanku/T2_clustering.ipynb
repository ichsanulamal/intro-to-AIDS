### Soal 1 [15 Poin] - Teori



#### a. Jelaskan mengapa K-means clustering tidak cocok untuk yang bukan hyper-spheres?


Algoritma K-means bekerja sedemikian sehingga setiap kali dua titik data yang berbeda memiliki jarak yang substansial satu sama lain (meskipun mereka adalah bagian dari kumpulan data yang sama), ia dapat menempatkan mereka ke dalam dua cluster yang berbeda setelah dipartisi. Dengan demikian, K-means cenderung membentuk cluster yang hyper-spheres, yang secara geometris adalah jarak antara setiap dua titik kecil. Tentunya hal itu kurang cocok diterapkan pada yang bukan hyper-spheres. Contoh kurang cocoknya penggunaan K-means clustering dapat dilihat pada nomor 3 bagian d.

#### b. Jelaskan mengapa algoritma Hierarchical clustering dapat digunakan tanpa perlu menetapkan jumlah cluster?

Algoritma hierarchical clustering dapat bekerja dengan cara agglomerative ataupun divisive. Apabila menggunakan divisive clustering, algoritma tersebut menginisialisasi tree yang mengandung semua titik data, kemudian setiap iterasi, data pada node spesifik dibagi menjadi beberapa beberapa node atau cluster. Sebaliknya, pada agglomerative clustering, algoritma dimulai dari setiap titik pada data, dan dari setiap data tersebut dapat dimerge menjadi cluster baru hingga menjadi hanya satu cluster. Karena kedua algoritma tersebut menyimpan tahapan clusterisasinya, kita tidak perlu menetapkan jumlah cluster di awal, tetapi kita perlu memilih jumlah cluster setelah proses hierarchical clustering selesai.





#### c. Jelaskan apa itu nilai metrik intra-class similarity dan cara menghitungnya!

Berdasarkan Silhouette Coefficient Formulations, Intra-class similarity merupakan rata-rata jarak antara satu objek dengan objek lainnya pada suatu cluster yang sama. Cara menghitungnya adalah dengan menjumlahkan jarak satu objek tersebut dengan objek lainnya pada cluster, kemudian membaginya dengan jumlah elemen pada cluster. Jarak tiap objek dapat dihitung melalui Complete Diameter Distance, Average Diameter Distance, atau Centroid Diameter Distance. Semakin kecil intra-class similarity, semakin baik pula clustering yang berarti elemen pada cluster sangat mirip.




#### d. Jelaskan apa itu nilai metrik inter-class similarity dan cara menghitungnya!

Berdasarkan Silhouette Coefficient Formulations, Inter-class similarity merupakan rata-rata jarak antara 1 objek pada satu cluster dengan seluruh objek pada cluster lainnya yang berbeda. Cara menghitungnya adalah dengan menjumlahkan setiap jarak 1 objek tersebut dengan objek lainnya di cluster berbeda, kemudian membaginya dengan banyak elemen pada cluster yang berbeda. Cara menghitung jaraknya bisa melalui Single Linkage Distance, Complete Linkage Distance, Average Linkage Distance, Centroid Linkage Distance, atau Average Centroid Linkage Distance. Semakin besar inter-class similarity semakin baik yang artinya terdapat perbedaan signifikan antara 2 cluster berbeda.

### Soal 2 [40 Poin] - Guess the clustering





#### a. Buka data dari soal2.csv dan plot semua titik di plot dua dimensi!

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
soal2 = pd.read_csv("soal2.csv")
sns.scatterplot(x="x1", y="x2", data=soal2)
```

#### b. Dari plot tersebut, usulkan jumlah klaster yang dapat membagi data dengan baik!



Saya mengusulkan 11 klaster untuk dapat membagi data dengan baik

#### c. Gunakan sebuah algoritma clustering yang dapat membagi data sesuai dengan jumlah klasteryang Anda usulkan. Jelaskan mengapa algoritma clustering tersebut digunakan!

Terlihat pada visualisasi bahwa tiap titik data tergabung pada suatu cluster yang jelas atau bersifat konvergen. Algoritma K-Means relatif sederhana untuk diimplementasikan dan dapat dengan mudah mengelompokkan data tersebut sehingga saya memilih menggunakannya.



```python
from sklearn.cluster import KMeans
```

```python
# Berikut merupakan Elbow Method untuk mengetahui jumlah cluster yang cocok untuk data soal2.csv
ks = range(5, 15)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(soal2)
    inertias.append(model.inertia_)

plt.figure(figsize=(8,5))
plt.style.use('bmh')
plt.plot(ks, inertias, '-o')
plt.xlabel('Number of clusters, k')
plt.ylabel('Inertia')
plt.xticks(ks)
plt.show()
```

```python
kmeans = KMeans(n_clusters=11)
pred_y = kmeans.fit_predict(soal2)
```

#### d. Visualisasikan hasil clustering dengan menampilkan scatter plot data yang di color-codedberdasarkan klasternya. Selain itu, cetak jumlah cluster yang dihasilkan!
**Hint**: Pelajari parameter c pada fungsi matplotlib.pyplot.scatter()



```python
sns.scatterplot(x="x1", y="x2", data=soal2, c=pred_y)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=50, c='red')
plt.show()
```

#### e. Hitung nilai intra-class similarity hasil clustering tersebut dengan menghitung jumlah darijarak (sum of distance) masing-masing sampel dengan pusat klasternya! Koordinat sebuah pusat klaster adalah rata-rata dari semua sampel di klaster tersebut. Rumus jarak yang digunakan adalah L2-norm/Euclidean Distance.



```python
res = soal2.copy()
res["y"] = pred_y

ics = []

print("Rata-rata Intra-class similarity untuk setiap cluster")

pd.DataFrame()

for center in pd.unique(pred_y):
    df_center = res[res["y"] == center]
    x1 = df_center["x1"].mean()
    x2 = df_center["x2"].mean()

    ics_i = (df_center[['x1', 'x2']] - np.array([x1, x2])).pow(2).sum(1).pow(0.5).sum() / (len(df_center)-1)
    ics.append([center, ics_i])

df_result = pd.DataFrame(np.array(ics), columns=["cluster_label", "ics"])
df_result
```

#### f. Hitung nilai silhouette coefficient dari hasil clustering tersebut!

```python
from sklearn.metrics import silhouette_score
print(f'Silhouette Score(n=11): {silhouette_score(soal2, pred_y)}')
```

#### a. Buka data dari soal4.csv dan tampilkan 10 baris pertama! Hitung jumlah data dan jumlah fitur!

```python
soal4 = pd.read_csv("soal4.csv")
soal4.head(10)
```

#### b. Ikuti langkah-langkah berikut:


##### i. Salin data asli dan masukkan dalam variable data_copy_1!

```python
data_copy_1 = soal4.copy()
```

##### ii. Lakukan Dimensionality reduction (PCA) pada data_copy_1 sehingga jumlah fiturnya menjadi 2!

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
skl_pca = pca.fit_transform(data_copy_1)
skl_pca.shape
```

##### iii. Visualisasikan data_copy_1 yang sudah direduksi, kemudian tentukan berapa jumlah cluster yang tepat!



```python
principalDf = pd.DataFrame(data = skl_pca
             , columns = ['principal component 1', 'principal component 2'])

fig = plt.figure(figsize = (8,8))

ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)

sns.scatterplot(x='principal component 1', y='principal component 2', 
                palette=sns.color_palette("hls", 3),
                cmap='CMRmap', 
                marker='o',
                data=principalDf, 
                )

ax.grid()
```

```python
ks = range(1, 10)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(principalDf)
    inertias.append(model.inertia_)

plt.figure(figsize=(8,5))
plt.style.use('bmh')
plt.plot(ks, inertias, '-o')
plt.xlabel('Number of clusters, k')
plt.ylabel('Inertia')
plt.xticks(ks)
plt.show()
```

Jumlah cluster yang tepat adalah sekitar 2 dan 3

##### iv. Lakukan clustering pada data_copy_1 yang sudah direduksi menggunakan K-Means dengan parameter random_state=2021!



```python
kmeans = KMeans(n_clusters=3, random_state=2021)
pred_y = kmeans.fit_predict(principalDf)
```

##### v. Visualisasikan hasil clustering pada data_copy_1 yang sudah direduksi!



```python
plt.scatter(principalDf["principal component 1"], principalDf["principal component 2"], c=pred_y)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=50, c='red')
plt.show()
```
